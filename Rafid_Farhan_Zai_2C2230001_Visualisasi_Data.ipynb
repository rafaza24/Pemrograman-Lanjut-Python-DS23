4.Mengambil Ekstremum


```python
# Membaca file CSV bernama "tingkatinflasi20082013.csv"
df = pd.read_csv("tingkatinflasi20082013.csv")

# Mengambil data dengan tahun 2012
print(df.loc[df["Tahun"] == 2012])
# Output:
# Tahun Cakupan Tingkat_Inflasi
# 12 2012 Kota Bandung 4.02
# 13 2012 Prov. Jawa Barat 3.86
# 14 2012 Nasional 4.30

# Mengambil data Kota Bandung sebelum tahun 2012
print(df.loc[(df["Cakupan"] == "Kota Bandung") & (df["Tahun"] < 2012)])
# Output:
# Tahun Cakupan Tingkat_Inflasi
# 0 2008 Kota Bandung 10.23
# 3 2009 Kota Bandung 2.11
# 6 2010 Kota Bandung 4.53
# 9 2011 Kota Bandung 2.75

# Mengambil data dengan tingkat inflasi di atas 10 atau di bawah 3
print(df.loc[(df["Tingkat_Inflasi"] > 10) | (df["Tingkat_Inflasi"] < 3)])
# Output:
# Tahun Cakupan Tingkat_Inflasi
# 0 2008 Kota Bandung 10.23
# 1 2008 Prov. Jawa Barat 11.11
# 2 2008 Nasional 11.06
# 3 2009 Kota Bandung 2.11
# 4 2009 Prov. Jawa Barat 2.02
# 5 2009 Nasional 2.78
# 9 2011 Kota Bandung 2.75

# Mengambil data dengan inflasi minimum
imin = df[" Tingkat_Inflasi"].idxmin()
print(df[imin: imin + 1])
# Output:
# Tahun Cakupan Tingkat_Inflasi
# 4 2009 Prov. Jawa Barat 2.02
```

5.Mengurutkan Data

# Bagian Baru

1.Histogram distribusi nilai matematika

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))
plt.hist(df['nilai_mat'], bins=10, color='cyan', edgecolor='black')
plt.title('Distribusi Nilai Matematika', fontsize=14)
plt.xlabel('Nilai Matematika', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Keterangan :
Grafik ini menunjukkan bahwa sebagian besar mahasiswa memiliki nilai matematika di rentang 60-80.

2.Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas.

```python
fakultas_count = df['fakultas'].value_counts()


plt.figure(figsize=(8, 6))
plt.barh(fakultas_count.index, fakultas_count.values, color='cyan', edgecolor='black')
plt.title('Jumlah Mahasiswa per Fakultas', fontsize=14)
plt.xlabel('Jumlah Mahasiswa', fontsize=12)
plt.ylabel('Fakultas', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```

Keterangan:
Fakultas dengan jumlah mahasiswa terbanyak adalah FTI.

3.Diagram pie banyaknya mahasiswa masing-masing fakultas

```python
plt.figure(figsize=(8, 8))
plt.pie(fakultas_count, labels=fakultas_count.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140)
plt.title('Proporsi Mahasiswa per Fakultas', fontsize=14)
plt.show()
```

Menampilkan proporsi mahasiswa di tiap fakultas dalam bentuk persentase.

4.Berdasarkan diagram pie dan batang horizontal manakah fakultas dengan mahasiswa terbanyak?

```python
mahasiswa_fakultas = fakultas_count.idxmax()
mahasiswa_fakultas
```

Fakultas dengan mahasiswa terbanyak: Fakultas Teknologi Industri (FTI).

Diagram batang lebih baik dalam menunjukkan jumlah mahasiswa per fakultas secara langsung, karena mempermudah perbandingan visual.
